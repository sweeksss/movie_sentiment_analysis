```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import re
```

```python
pwd = os.getcwd()
```

```python
# Getting the filepath
file_path_train_pos = (pwd + "//aclImdb//train//pos")
```

Dataframe creation starts

```python
# Function to read the content of text files
def read_file_content(file_path):
    with open(file_path, 'r', errors='ignore') as file:
        content = file.read()
    return content
```

```python
# List to store the data
data_train_pos = []
```

```python
# Extracting information from the text files
for filename in (os.listdir(file_path_train_pos)):
    if filename.endswith(".txt"):
        movie_id, rating = filename.split("_")[:2]
        content = read_file_content(os.path.join(file_path_train_pos, filename))
        data_train_pos.append({'id': int(movie_id), 'movie_rating': rating, 'review': content, 'label': 'good'})

# Convert list of dictionaries to DataFrame
df_train_pos = pd.DataFrame(data_train_pos)

# Sort DataFrame by 'id'
df_train_pos = df_train_pos.sort_values(by='id')
```

```python
df_train_pos
```

```python
# Reset index starting from 0
df_train_pos = df_train_pos.reset_index(drop=True)
df_train_pos
```

```python
file_path_urls_pos = (pwd + "//aclImdb//train//urls_pos.txt")
```

```python
# Function to extract movie id from the urls
def extract_id(url):
    return url.replace('/usercomments', '')
```

```python
# Extract movie id from urls
pattern = r'tt\d+'
movie_identifier = []
movie_urls = []
with open(file_path_urls_pos, 'r') as file:
    imdb_links = file.readlines()

for link in imdb_links:
    url = extract_id(link.strip())
    movie_urls.append(url)

print(movie_urls)

for link in movie_urls:
    mov_id = link.split('/')
    movie_identifier.append(mov_id[4])

print(movie_identifier)
```

```python
# Add a movie id column
df_train_pos['movie_id'] = movie_identifier

# Define the column to move and its new position
column_to_move = 'movie_id'
new_position = 1  # 0-indexed position

# Get the list of columns
columns = df_train_pos.columns.tolist()

# Remove the column to move from its original position
columns.remove(column_to_move)

# Insert the column at the new position
columns.insert(new_position, column_to_move)

# Reorder the DataFrame columns
df_train_pos = df_train_pos[columns]

# Drop id column to use movie_id as the actual movie identifier
df_train_pos.drop('id', axis=1, inplace=True)

df_train_pos
```

```python
# Getting rid of ".txt" from movie_rating column and changing datatype to int
df_train_pos['movie_rating'] = df_train_pos['movie_rating'].str.replace('.txt', '').astype(int)

# Final dataframe with negative movie reviews
df_train_pos
```
